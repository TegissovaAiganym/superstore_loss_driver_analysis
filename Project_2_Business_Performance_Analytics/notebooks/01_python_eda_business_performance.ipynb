```python
#changing column names to standard format in lowercase, underscores only
sales_data.columns = (
    sales_data.columns
    .str.lower()
    .str.replace(' ', '_' )
    .str.replace('-', '_')
)
```

```python
#convert order and ship dates to datetime
sales_data['order_date'] = pd.to_datetime(sales_data['order_date'])
sales_data['ship_date'] = pd.to_datetime(sales_data['ship_date'])
sales_data[['order_date','ship_date']].dtypes
```

```python
#KPI analysis
# calculate profit margin by dividing profit to sales
sales_data['profit_margin'] = sales_data['profit']/sales_data['sales']
```

```python
#calculate difference in days between order date and ship date. It will help us to see any delays 
sales_data['delivery_days'] = sales_data['ship_date'] - sales_data['order_date']
```

```python
sales_data['delivery_days'] = sales_data['ship_date'] - sales_data['order_date']#Business question 1: Which regions generate the most sales and profit
region_analysis = (
    sales_data
    .groupby('region')[['sales', 'profit']] #split dataset into groups according to regions, only show sales and profit
    .sum() #add up sales, profit values for each region
    .sort_values(by='sales', ascending = False) 
)
region_analysis
```

```python
#Business question 2: Which products generate the most revenue and profit

category_analysis = (
    sales_data
    .groupby('category')[['sales', 'profit']]
    .sum()
)
category_analysis
```

```python
#Business question 2: Which categories are more profitable relative to their sales?
category_profitability = (
    sales_data
    .groupby('category')[['profit_margin']]
    .mean()
)
category_profitability
# this is the calculation of the unweighted average, so small and large orders have the same impact
#it is good for understanding per-order behaviour, comparing pricing efficiency
```

```python
#Calculation of weighted profit margin by dividing total profit by total sales
#Categories with higher sales have more influence on the result (reflects real business profitability)
category_profitability_weighted = category_analysis['profit'] / category_analysis['sales']
data = pd.DataFrame(category_profitability_weighted, columns = ['weighted_profit_margin'])
data
```

```python
# Are specific shipping methods are faster than others on average? 
time_by_ship_mode = (
    sales_data
    .groupby('ship_mode')['delivery_days']
    .mean()
)
time_by_ship_mode
```

```python
# Does delivery speed impact profitability? 
# Order grouped by delivery speed to compare average profit between fast and slow deliveries
fast_mask = sales_data['delivery_days'] <= pd.Timedelta(days = 1)
medium_mask = ( sales_data['delivery_days'] > pd.Timedelta(days = 1) ) & ( sales_data['delivery_days'] <= pd.Timedelta(days = 3) )
slow_mask = sales_data['delivery_days'] >= pd.Timedelta(days = 4)

sales_data[medium_mask]
```

```python
sales_data['delivery_speed']= 'Unknown'
sales_data['delivery_speed'].value_counts(dropna = False)
#groupby drop NaN by default later in the code, so assigned everything to 'Unknown' first
```

```python

sales_data.loc[fast_mask,'delivery_speed'] = 'Fast'
sales_data.loc[medium_mask,'delivery_speed'] = 'Medium'
sales_data.loc[slow_mask,'delivery_speed'] = 'Slow'
sales_data[:10]
```

```python
profit_by_delivery = (
    sales_data
    .groupby('delivery_speed')['profit_margin']
    .mean()
)
profit_by_delivery
```
